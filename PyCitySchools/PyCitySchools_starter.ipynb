```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate total number of schools and students
total_schools = len(school_data_complete['School ID'].unique())
total_students = len(school_data_complete['student_name'])

#Calculate the total budget
school_budget = []
school_budget.append(school_data_complete['budget'].unique())
total_budget = round(np.sum(school_budget)/1000000)

#Calculate average math and reading scores 
ave_math = round(school_data_complete['math_score'].mean())
ave_reading = round(school_data_complete['reading_score'].mean())

#Calculate %passed for math, reading, math & reading
passed_math = school_data_complete.loc[(school_data_complete['math_score']>=70)]
passed_math_percent = round((len(passed_math) / total_students) *100)
passed_reading = school_data_complete.loc[(school_data_complete['reading_score']>=70)]
passed_reading_percent = round((len(passed_reading) / total_students) *100)
passed_both = school_data_complete.loc[(school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70)]
passed_both_percent = round((len(passed_both) / total_students) *100)

#New dataframe to hold results above
district_summary = pd.DataFrame({"Total # of Schools": [total_schools], "Total # of Students": [total_students], "Total Budget (millions)": [total_budget], "Average Math Score": [ave_math], "Average Reading Score": [ave_reading], "% Passed Math":[passed_math_percent],"% Passed Reading":[passed_reading_percent],"% Passed Math & Reading":[passed_both_percent]})
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Group dataframe by school
school_data_complete.rename(columns = {"type":"School Type"})
school_grouped = school_data_complete.groupby(['school_name'])

#Identify school type and school budget
school_type = school_grouped["type"].nth(1)
budget = school_grouped["budget"].nth(1)

#Calculate total number of students in each school
total_students2 = school_grouped["student_name"].size()

#Calculate total budget per student for each school
budget_student = round((school_grouped["budget"].nth(1)) / school_grouped["student_name"].size())

#Calculate average math and reading scores for each school
ave_mathscore = round(school_grouped["math_score"].mean())
ave_readingscore = round(school_grouped["reading_score"].mean())

#Calculate % passing math
passed_math = school_data_complete.loc[school_data_complete['math_score']>=70]
passed_math_grouped = passed_math.groupby(['school_name'])
number_passed_math = passed_math_grouped['student_name'].count()
percent_passed_math = round((number_passed_math / total_students2) *100)

#Calculate % passing reading
passed_reading = school_data_complete.loc[school_data_complete['reading_score']>=70]
passed_reading_grouped = passed_reading.groupby(['school_name'])
number_passed_reading = passed_reading_grouped['student_name'].count()
percent_passed_reading = round((number_passed_reading / total_students2) *100)

#Calculate % overall passing
passed_both = school_data_complete.loc[(school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70)]
passed_both_grouped = passed_both.groupby(['school_name'])
number_passed_both = passed_both_grouped['student_name'].count()
percent_passed_overall = (number_passed_both / total_students2) *100

#New dataframe to hold results above
school_summary = pd.DataFrame(school_type)
school_summary = school_summary.set_axis(["School Type"],axis=1)
school_summary.index.name = None
school_summary['Total Students'] = total_students2
school_summary['Total School Budget'] = budget
school_summary['Per Student Budget'] = budget_student
school_summary['Average Math Score'] = ave_mathscore
school_summary['Average Reading Score'] = ave_readingscore
school_summary['% Passing Math'] = percent_passed_math
school_summary['% Passing Reading'] = percent_passed_reading
school_summary['% Overall Passing'] = percent_passed_overall
school_summary
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Create a pandas series for each grade
mathdf = school_data_complete[["school_name","grade","math_score"]]
mathdf.head()

grade9 = mathdf.loc[mathdf["grade"] == "9th", :]
grade10 = mathdf.loc[mathdf["grade"] == "10th", :]
grade11 = mathdf.loc[mathdf["grade"] == "11th", :]
grade12 = mathdf.loc[mathdf["grade"] == "12th", :]

#Group each series by school
grade9_grouped = grade9.groupby(['school_name']).mean()
grade10_grouped = grade10.groupby(['school_name']).mean()
grade11_grouped = grade11.groupby(['school_name']).mean()
grade12_grouped = grade12.groupby(['school_name']).mean()

#Combine the series into a dataframe
summary = pd.merge(grade9_grouped, grade10_grouped, on="school_name", suffixes=("_9", "_10"))
summary2 = pd.merge(summary, grade11_grouped, on="school_name")
summary3 = pd.merge(summary2, grade12_grouped, on="school_name")
summary_table = summary3.rename(columns = {"math_score_9": "9th", "math_score_10": "10th", "math_score_x": "11th", "math_score_y": "12th"})
summary_table.index.name = None
summary_table
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Cut columns not needed in complete dataframe 
school_data = school_data_complete[['school_name','reading_score','math_score','budget','student_name']]

#Sort dataframe by school name and calculate budget per student
school = school_data.groupby(['school_name'])
total_students = school['student_name'].count()
school_budget = school['budget'].nth(1)
perstudent = school_budget / total_students
perstudent_df = pd.DataFrame(perstudent)
perstudent_df = perstudent_df.set_axis(['Budget (per student)'],axis=1)

#Merge budget per student dataframe with original dataframe to assign each school a "per student" budget
add_perstudent = pd.merge(school_data, perstudent_df, on = 'school_name')

#Cut the merged dataframe using bins
bins = [0,585,630,645,680]
group_names = ["<$585", "$585-630", "$630-645","$645-680"]
add_perstudent['Spending Ranges (Per Student)'] = pd.cut(add_perstudent['Budget (per student)'],bins, labels = group_names, include_lowest = True)

#Calculate average math and reading scores per bin
grouped_df = add_perstudent.groupby(['Spending Ranges (Per Student)'])
average_math_binned = round(grouped_df['math_score'].mean())
average_reading_binned = round(grouped_df['reading_score'].mean())
total_students_binned = grouped_df['student_name'].size()

#Calculate passing math scores per bin 
passed_math_bins = add_perstudent.loc[(add_perstudent['math_score']>=70)]
grouped_math = passed_math_bins.groupby(['Spending Ranges (Per Student)'])
passed_math_binned = round((grouped_math['math_score'].size() / total_students_binned)*100)

#Calculate passing reading scores per bin 
passed_reading_bins = add_perstudent.loc[(add_perstudent['reading_score']>=70)]
grouped_reading = passed_reading_bins.groupby(['Spending Ranges (Per Student)'])
passed_reading_binned = round((grouped_reading['reading_score'].size() / total_students_binned)*100)

#Calculate overall passing 
passed_overall_bins = add_perstudent.loc[(add_perstudent['math_score']>=70) & (add_perstudent['reading_score']>=70)]
grouped_overall = passed_overall_bins.groupby(['Spending Ranges (Per Student)'])
passed_overall_binned = round((grouped_overall['student_name'].size() / total_students_binned)*100)

#Summary table
budget_summary = pd.merge(average_math_binned, average_reading_binned, on='Spending Ranges (Per Student)')
budget_table = budget_summary.rename(columns={'math_score': 'Average Math Score', 'reading_score': "Average Reading Score"})
budget_table["% Passing Math"] = passed_math_binned
budget_table["% Passing Reading"] = passed_reading_binned
budget_table["% Overall Passing"] = passed_overall_binned
budget_table
average_math_binned
```

```python
#Cut columns not needed in complete dataframe 
size = school_data_complete[['school_name','reading_score','math_score','budget','student_name']]

#Sort dataframe by school name and calculate school size
size_grouped = size.groupby(['school_name'])
school_size = size_grouped['student_name'].size()
size_df = pd.DataFrame(school_size)
size_df = size_df.set_axis(['School Size'],axis=1)
size_df

#Merge dataframes
schoolsize = pd.merge(school_data_complete, size_df, on = 'school_name')

#Cut the merged dataframe using bins
bins = [0,1000,2000,3000]
group_names = ["Small <1000", "Medium (1000-2000)", "Large (2000-5000)"]
schoolsize['School Size'] = pd.cut(schoolsize['School Size'],bins, labels = group_names, include_lowest = True)

#Calculate average math and reading scores per bin
schoolsize_grouped = schoolsize.groupby(['School Size'])
ave_math = round(schoolsize_grouped['math_score'].mean())
ave_reading = round(schoolsize_grouped['reading_score'].mean())
total_students = schoolsize_grouped['student_name'].size()

#Calculate passing math scores per bin 
passed_math = schoolsize.loc[(schoolsize['math_score']>=70)]
passed_math_grouped = passed_math.groupby(['School Size'])
passed_math_total = passed_math_grouped['math_score'].size()
passed_math_percent = round((passed_math_total / total_students) *100)

#Calculate passing readingh scores per bin 
passed_reading = schoolsize.loc[(schoolsize['reading_score']>=70)]
passed_reading = passed_reading.groupby(['School Size'])
passed_reading = passed_reading['reading_score'].size()
passed_reading_percent = round((passed_reading / total_students) *100)

#Calculate overall passing per bin 
passed_both = schoolsize.loc[(schoolsize['math_score']>=70) & (schoolsize['reading_score']>=70)]
passed_both = passed_both.groupby(['School Size'])
passed_both = passed_both['math_score'].size()
passed_both_percent = round((passed_both / total_students) *100)

#Summary table
budget_summary = pd.merge(ave_math, ave_reading, on='School Size')
budget_table = budget_summary.rename(columns={'math_score': 'Average Math Score', 'reading_score': "Average Reading Score"})
budget_table["% Passing Math"] = passed_math_percent
budget_table["% Passing Reading"] = passed_reading_percent
budget_table["% Overall Passing"] = passed_both_percent
budget_table

```
